```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
 CBB = pd.read_csv('Data\cbb.csv')
```

```python
CBB
```

```python
CBB.dtypes
```

```python
CBB.describe()
```

```python
my_stats = ["TEAM", "CONF", "W","ADJOE", "ADJDE", "SEED", "POSTSEASON"]
cbb = CBB[my_stats]
```

```python
cbb = cbb.dropna()
```

```python
cbb["POSTSEASON"].value_counts()
```

```python
sns.set_style("darkgrid")
```

```python
sns.displot(data= cbb, x = "SEED", y = "POSTSEASON")
```

```python
sns.displot(data= cbb, x = "ADJOE", y = "POSTSEASON")
```

```python
seed_plot = sns.violinplot(x='POSTSEASON', y='SEED', data=cbb, order=["R68", "R64","R32","S16", "E8", "F4", "2ND","Champions"])
seed_plot.set(ylim = (1,16))
seed_plot.set(
    title= "Postseason results by March Madness Seed")
plt.savefig("SEEDPOSTSEASON.png")
```

```python
ps = sns.scatterplot(data=cbb, x="ADJOE", y="ADJDE", hue = "POSTSEASON")
ps.set(title = "Postseason results by ADJOE and ADJDE")
plt.savefig("ADJOE_VS_ADJED_PS.png")
```

```python
cbb_BigSky = CBB[(CBB["CONF"] == "BSky")]
my_stats = ["TEAM", "CONF", "W","ADJOE", "ADJDE","YEAR", "SEED", "POSTSEASON"]
cbb_BigSky = cbb_BigSky[my_stats]
cbb_BigSky['POSTSEASON'] = cbb_BigSky['POSTSEASON'].fillna("NONE")
cbb_BigSky['SEED'] = cbb_BigSky['SEED'].fillna("NONE")
##cbb_BigSky = cbb_BigSky.dropna()
cbb_BigSky
```

```python
sns.scatterplot(data=cbb_BigSky, x="ADJOE", y="ADJDE", hue = "SEED").set(title = "Big Sky Seed by ADJOE and ADJDE")
plt.savefig("BIGSKY_ADJOE_VS_ADJED_SEED.png")
```
